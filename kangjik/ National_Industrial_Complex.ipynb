```python
import pandas as pd
import numpy as np

# excel 데이터 불러오기,
tmp = pd.read_excel('data/국가산업단지_산업동향_18_01.xlsx',sheet_name=9, skiprows=2)
tmp
```

```python
# 다 국가 산업단지이기 때문에 "구분" 칼럼 삭제하기
tmp.drop(["구분"], axis=1, inplace=True)
tmp
```

```python
# 산업단지 명을 인덱스로 사용하기
tmp.set_index("산업단지", inplace=True)
tmp
```

```python
# 계가 0인 산업단지 없애기
tmp = tmp[tmp.계 != 0]
tmp
```

```python
# 'x' 값을 가지는 행 삭제하기
tmp = tmp.replace('x', np.nan).dropna()
tmp
```

```python
# 계 칼럼 삭제하기
tmp_without_total = tmp.drop(["계"], axis=1)
tmp_without_total
```

```python
# 각 산업단지 별 제일 고용인구가 많은 업종
most = tmp_without_total.idxmax(axis=1)
most
```

```python
# 각 산업단지 별 제일 고용인구가 많은 업종의 종목명
most = tmp_without_total.max(axis=1)
most
```

```python
# 각 산업단지 별 제일 고용인구가 많은 종목의 차지비율
percentage = most / tmp_without_total.sum(axis=1) * 100
percentage
```

```python
# 이후 불러올 데이터들 위의 과정을 하면서 불러오기
tmp2 = pd.read_excel('data/국가산업단지_산업동향_18_02.xlsx',sheet_name=9, skiprows=2)
tmp2 = tmp2[tmp2.계 != 0.0]
tmp2 = tmp2.replace('x', np.nan).dropna()
tmp2.drop(["구분", '계'], axis=1, inplace=True) 
tmp2.set_index("산업단지", inplace=True)
tmp2
```

```python
tmp_set = set(tmp.산업단지.values)
tmp2_set = set(tmp2.산업단지.values)
print('22년 자료에는 있고, 18년엔 없음', tmp2_set - tmp_set)
print('18년 자료에는 있고, 22년엔 없음', tmp_set - tmp2_set)
```
