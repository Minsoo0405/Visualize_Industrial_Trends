```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 확인
tmp = pd.read_excel('../data/국가산업단지_산업동향_22_07.xlsx',sheet_name=9)
tmp
```

```python

# 데이터 전처리
df = pd.read_excel('../data/국가산업단지_산업동향_22_07.xlsx',sheet_name=9, skiprows=2)
df = df[df.계 != 0.0]
df = df.replace('X', np.nan).dropna()
df.drop(["구분"], axis=1, inplace=True) 
df.set_index("산업단지", inplace=True)
df
```

### 총계를 기준으로 정렬 진행

```python
df.sort_values(by=['계'], ascending=False, inplace= True)
df
```

### 총 고용인구가 20000명 이상인 곳만 다루기 위해 데이터프레임 추출

```python
df2 = df[df['계'] >= 20000]
df2
```

### 파이차트 시각화를 위해 계 column 제거

```python
df2 = df2.drop(['계'], axis=1)
df2
```

### 시각화

```python
labels = df2.columns
explode = [0.05] * len(labels)
indices = df2.index
colors=['black', 'lightpink', 'darkgray', 'hotpink', 'rosybrown', 'deeppink',
        'darkorange', 'blue', 'lime', 'cyan', 'teal']
```

```python
# Index(['서울', '시화', '창원', '반월', '울산ㆍ미포', '남동', '구미', '녹산', '여수'], dtype='object', name='산업단지')
subplot_counter = 1
for index in indices:
    plot_tmp_df = df2.loc[index]
    values = plot_tmp_df.values
    plt.subplot(3, 3, subplot_counter)
    plt.title(index)
    plt.pie(values, autopct='%1.1f%%', explode=explode, colors=colors)
    subplot_counter += 1

plt.legend(labels)
plt.tight_layout()
plt.show()
```

```python
# subplot 말고 하나씩 plot하고 저장
for index in indices:
    plot_tmp_df = df2.loc[index]
    values = plot_tmp_df.values
    plt.title(index + "업종별 고용 인구 비율")
    plt.pie(values, autopct='%1.1f%%', explode=explode, colors=colors)
    plt.legend(labels)
    plt.savefig(index + '.png')
    plt.show()        
```
