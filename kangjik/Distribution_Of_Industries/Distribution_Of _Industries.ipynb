```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 확인
tmp = pd.read_excel('../data/국가산업단지_산업동향_22_07.xlsx',sheet_name=9)
tmp
```

```python

# 데이터 전처리
df = pd.read_excel('../data/국가산업단지_산업동향_22_07.xlsx',sheet_name=9, skiprows=2)
df = df[df.계 != 0.0]
df = df.replace('X', np.nan).dropna()
df.drop(["구분"], axis=1, inplace=True) 
df.set_index("산업단지", inplace=True)
df
```

### 총계를 기준으로 정렬 진행

```python
df.sort_values(by=['계'], ascending=False, inplace= True)
df
```

### 총 고용인구가 20000명 이상인 곳만 다루기 위해 데이터프레임 추출

```python
df2 = df[df['계'] >= 20000]
df2
```

### 파이차트 시각화를 위해 계 column 제거

```python
df2 = df2.drop(['계'], axis=1)
df2
```

```python
# Index(['서울', '시화', '창원', '반월', '울산ㆍ미포', '남동', '구미', '녹산', '여수'], dtype='object', name='산업단지')
labels = df2.columns
explode = [0.05] * len(labels)

# 서울
seoul = df2.loc["서울"]
values = seoul.values

plt.subplot(3, 3, 1)
plt.title("서울")
plt.pie(values, autopct='%1.1f%%', explode=explode)

# 시화
sihwa = df2.loc["시화"]
values = sihwa.values

plt.subplot(3, 3, 2)
plt.title("시화")
plt.pie(values, autopct='%1.1f%%', explode=explode)

# 창원
changwon = df2.loc["창원"]
values = changwon.values

plt.subplot(3, 3, 3)
plt.title("창원")
plt.pie(values, autopct='%1.1f%%', explode=explode)

# 반월
banwol = df2.loc["반월"]
values = banwol.values

plt.subplot(3, 3, 4)
plt.title("반월")
plt.pie(values, autopct='%1.1f%%', explode=explode)

# 울산 미포
ulsan_mipo = df2.loc['울산ㆍ미포']
values = ulsan_mipo.values

plt.subplot(3, 3, 5)
plt.title('울산ㆍ미포')
plt.pie(values, autopct='%1.1f%%', explode=explode)

# 남동
namdong = df2.loc['남동']
values = namdong.values

plt.subplot(3, 3, 6)
plt.title('남동')
plt.pie(values, autopct='%1.1f%%', explode=explode)

# 구미
gumi = df2.loc['구미']
values = gumi.values

plt.subplot(3, 3, 7)
plt.title('구미')
plt.pie(values, autopct='%1.1f%%', explode=explode)

# 녹산
noksan = df2.loc['녹산']
values = noksan.values

plt.subplot(3, 3, 8)
plt.title('녹산')
plt.pie(values, autopct='%1.1f%%', explode=explode)

# 여수
yeosu = df2.loc['여수']
values = yeosu.values

plt.subplot(3, 3, 9)
plt.title('여수')
plt.pie(values, autopct='%1.1f%%', explode=explode)



plt.legend(labels)
plt.tight_layout()
plt.show()
```
