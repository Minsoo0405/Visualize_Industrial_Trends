```python
import numpy as np
import pandas as pd
```

# 현재 인력 현황

### 업종 중분류 및 소분류 컬럼명 수정

```python
xlsx_status = pd.read_excel('../data/data_ids_jobmanpower.xlsx')

xlsx_status.rename(columns={
                xlsx_status.columns[0] : "업종 중분류",
                xlsx_status.columns[1] : "업종 소분류"}, inplace=True)
```

```python
xlsx_status
```

### 연도별로 데이터프레임으로 쪼개기

```python
xlsx_status_2016 = xlsx_status[['업종 중분류', '업종 소분류', '2016년']]
xlsx_status_2017 = xlsx_status[['업종 중분류', '업종 소분류', '2017년']]
xlsx_status_2018 = xlsx_status[['업종 중분류', '업종 소분류', '2018년']]
xlsx_status_2019 = xlsx_status[['업종 중분류', '업종 소분류', '2019년']]
xlsx_status_2020 = xlsx_status[['업종 중분류', '업종 소분류', '2020년']]
xlsx_status_2021 = xlsx_status[['업종 중분류', '업종 소분류', '2021년']]
```

### 이전 컬럼(연도)를 인력현황로 바꾸기

```python
xlsx_status_2016.columns = ['업종 중분류', '업종 소분류', '현재인력현황']
xlsx_status_2017.columns = ['업종 중분류', '업종 소분류', '현재인력현황']
xlsx_status_2018.columns = ['업종 중분류', '업종 소분류', '현재인력현황']
xlsx_status_2019.columns = ['업종 중분류', '업종 소분류', '현재인력현황']
xlsx_status_2020.columns = ['업종 중분류', '업종 소분류', '현재인력현황']
xlsx_status_2021.columns = ['업종 중분류', '업종 소분류', '현재인력현황']
```

### 하나의 데이터프레임으로 다시 합치기

```python
new_xlsx_status = pd.concat([xlsx_status_2016, xlsx_status_2017, xlsx_status_2018, xlsx_status_2019, xlsx_status_2020, xlsx_status_2021], axis=0)
```

```python
new_xlsx_status['연도'] = 0

new_xlsx_status['연도'][0:24] = 2016
new_xlsx_status['연도'][24:48] = 2017
new_xlsx_status['연도'][48:72] = 2018
new_xlsx_status['연도'][72:96] = 2019
new_xlsx_status['연도'][96:120] = 2020
new_xlsx_status['연도'][120:144] = 2021
```

```python
new_xlsx_status
```

# 필요인력현황

### 업종 중분류 및 소분류 컬럼명 수정

```python
xlsx_nece = pd.read_excel('../data/data_ids_necemanpower.xlsx')

xlsx_nece.rename(columns={
                xlsx_nece.columns[0] : "업종 중분류",
                xlsx_nece.columns[1] : "업종 소분류"}, inplace=True)
```

### 하나의 데이터프레임으로 다시 합치기

```python
new_nece_xlsx = pd.concat([xlsx_nece_2019, xlsx_nece_2020, xlsx_nece_2021, xlsx_nece_2022], axis=0)
```

```python
new_nece_xlsx['연도'] = 0

new_nece_xlsx['연도'][0:24] = 2019
new_nece_xlsx['연도'][24:48] = 2020
new_nece_xlsx['연도'][48:72] = 2021
new_nece_xlsx['연도'][72:96] = 2022
```

```python
new_nece_xlsx
```

### 현재 인력현황과 필요인력현황 데이터프레임 합치기
